```python
f = open('wifi.csv', encoding = 'euc-kr')
fd = csv.reader(f)

mymap = folium.Map(location = [37.4980, 127.0275], zoom_start = 13)
next(fd)

q = input('어느 지역 무료 와이파이를 보고 싶으신가요? : ')
for r in fd:
  if q in r[9]:
    folium.Marker([float(r[12]), float(r[13])], popup = r[0]).add_to(mymap)

mymap
```
